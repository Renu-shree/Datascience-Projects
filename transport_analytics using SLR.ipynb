### DATA DESCRIPTION: 

The New York City Department of Transportation collects daily data about the number of bicycles going over bridges in New York City. This data is used to measure bike utilization as a part of transportation planning. This dataset is a daily record of the number of bicycles crossing into or out of Manhattan via one of the East River bridges (that is, excluding Bronx thruways and the non-bikeable Hudson River tunnels) for a stretch of 9 months.

```python
from sklearn.model_selection  import train_test_split, cross_val_score
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

X=df.drop(columns=["Total"])
y=df["Total"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)

```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

lr_model=LinearRegression()
selector=RFE(lr_model,n_features_to_select=5)
selector.fit(X_train,y_train)

selected_features=X.columns[selector.support_]
```

```python
X_train_selected=X_train[selected_features]
X_test_selected=X_test[selected_features]
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train_selected)
X_test_scaled=scaler.fit_transform(X_test_selected)
```

```python
ridge_model=Ridge(alpha=1.0)
ridge_model.fit(X_train_scaled,y_train)
```

```python
train_rmse=np.sqrt(mean_squared_error(y_train,ridge_model.predict(X_train_scaled)))
train_rmse
```

```python
test_rmse=np.sqrt(mean_squared_error(y_test,ridge_model.predict(X_test_scaled)))
test_rmse
```

```python
cv_scores=cross_val_score(ridge_model,X_train_scaled,y_train,cv=5,scoring="neg_mean_squared_error")
cv_rmse_scores=np.sqrt(-cv_scores)
print("cross validation score:",cv_rmse_scores)
print("mean cross validation score:",np.mean(cv_rmse_scores))
```

 The train rmse is low indiacting that model fits well on training data

the test rmse is significantly higher than train rmse indicating that model may not generalise well to unseen data.

the cross vaidation rmse scores are consistent across fols with mean cross vaidation rmse being close to test rmse.
This indicates that model performance is consistent across different subsets of training data

We can conclude that the model has reduced the overfitting compared to previous model but still there a chance for imporovemnt

```python
from sklearn.model_selection  import train_test_split, cross_val_score
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

X=df.drop(columns=["Total"])
y=df["Total"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import RidgeCV
ridge_cv_model=RidgeCV(alphas=np.logspace(-4,4,10),cv=5)
ridge_cv_model.fit(X_train_scaled,y_train)
```

```python
from sklearn.feature_selection import SelectFromModel
sfm=SelectFromModel(ridge_cv_model,threshold=0.25)
sfm.fit(X_train_scaled,y_train)
selected_features=X.columns[sfm.get_support()]
```

```python
X_train_selected=X_train[selected_features]
X_test_selected=X_test[selected_features]
```

```python
ridge_model=RidgeCV(alphas=np.logspace(-4,4,10),cv=5)
ridge_model.fit(X_train_selected,y_train)
```

```python
train_rmse=np.sqrt(mean_squared_error(y_train,ridge_model.predict(X_train_selected)))
train_rmse
```

```python
test_rmse=np.sqrt(mean_squared_error(y_test,ridge_model.predict(X_test_selected)))
test_rmse
```

```python
cv_scores=cross_val_score(ridge_model,X_train_selected,y_train,cv=5,scoring="neg_mean_squared_error")
cv_rmse_scores=np.sqrt(-cv_scores)
print("cross validation score:",cv_rmse_scores)
print("mean cross validation score:",np.mean(cv_rmse_scores))
```

Here we can see that train rmse and test rmse are very close to each other indicating that the model generalises well to unseen data

overall model performs resonabally well with low rmse indicating good predictive performance
